```python
def classify_transfer_type(value):
    if pd.isna(value):
        return 'Unknown'
    v = str(value).lower().strip()
    
    # Check for loan
    if 'loan' in v:  
        return 'Loan'
    # Check for free or zero cost
    if 'free' in v or v == '0' or v == '0.0' or v == '0€':
        return 'Free'
    # Otherwise permanent
    return 'Permanent'

# Apply the function to your DataFrame
summer_22['transfer_type'] = summer_22['cost'].apply(classify_transfer_type)

# Inspect the new column
print(summer_22[['cost','transfer_type']].head(15))

# (Optional) Save to a new CSV
summer_22.to_csv('summer_22_with_transfer_type.csv', index=False)
```

```python
import re


def convert_cost(value):
    if pd.isna(value) or value == "" or value == "?":
        return 0  

    value = str(value).lower().strip()

   
    if "free" in value:
        return 0

  
    if "m" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000_000) if num else 0

   
    if "th" in value or "k" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000) if num else 0

    
    if "loan" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000_000) if num else 0

    
    num = re.findall(r"[\d\.]+", value)
    return int(float(num[0])) if num else 0


summer_22["cost"] = summer_22["cost"].apply(convert_cost)


summer_22.to_csv("summer_market_clean.csv", index=False)

summer_22
```

```python
def convert_value(value):
    if pd.isna(value) or value == "" or value == "?":
        return 0  

    value = str(value).lower().strip()

   
    if "free" in value:
        return 0


    if "m" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000_000) if num else 0


    if "th" in value or "k" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000) if num else 0

    
    if "loan" in value:
        num = re.findall(r"[\d\.]+", value)
        return int(float(num[0]) * 1_000_000) if num else 0

    
    num = re.findall(r"[\d\.]+", value)
    return int(float(num[0])) if num else 0

summer_22["player_valuje"] = summer_22["player_valuje"].apply(convert_value)


summer_22.to_csv("summer_market_clean.csv", index=False)
summer_22
```
